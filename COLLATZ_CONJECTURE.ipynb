```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def collatz(x):
    sequence = [x]
    while x != 1:
        if x % 2 == 0:
            x = x/2
        else:
            x = 3*x +1
        sequence.append(x)
    return sequence
#        print(x)
        
```

```python
collatz(3)
```

```python
for x in range(1,16):
    print(collatz(x))
```

```python
x = 257
plt.plot(collatz(x));
```

```python
x = 257
plt.plot(np.log(collatz(x)));  # this looks like stock market trend
```

### Alternative codes for COLLATZ CONJECTURE

```python
def collatz(n):
    if n % 2 == 0: return int(n/2)
    else:          return 3*n+1

def collatz_sequence(n):
    sequence = [n]
    while n != 1:
        n = collatz(n)
        sequence += [n]
    return sequence
```

```python
collatz_sequence(7)
```

```python
plt.plot(collatz_sequence(464));
```

```python
plt.figure(figsize=(10,6))
#plt.plot([log2(i) for i in collatz_sequence(46463564)]);
#matplotlib don't work with only log, you need numpy i.e np.log2

plt.plot([np.log2(i) for i in collatz_sequence(46463564)]);
```

```python
#The sequences have a lot of random properties. 
# Plotting the length of collatz
plt.figure(figsize=(17,6))
plt.plot([len(collatz_sequence(i)) for i in range(1, 2**11)]);
```

```python
# What is the biggest value the sequence gets to, on a log plot:

plt.figure(figsize=(17,6))
plt.plot([np.log2(max(collatz_sequence(i))) for i in range(1, 2**11)]);
```
